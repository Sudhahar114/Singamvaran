```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
from pyspark.sql import Row

# Create Spark session
spark = SparkSession.builder.appName("SalesData").getOrCreate()

# Sample sales data with duplicates and nulls
sales_data = [
    (1, "Laptop", 55000.0, "2025-01-01"),
    (2, "Mouse", 500.0, "2025-01-02"),
    (3, "Keyboard", 1500.0, "2025-01-03"),
    (4, "Monitor", 12000.0, "2025-01-04"),
    (5, "Printer", 8000.0, "2025-01-05"),
    (6, "Tablet", 25000.0, "2025-01-06"),
    (7, "Laptop", 55000.0, "2025-01-07"),
    (8, "Mouse", None, "2025-01-08"),          # Null amount
    (9, None, 2000.0, "2025-01-09"),           # Null item
    (10, "Speaker", 3000.0, None),             # Null sale_date
    (11, "Camera", 45000.0, "2025-01-11"),
    (12, "Phone", 35000.0, "2025-01-12"),
    (13, "Charger", 1200.0, "2025-01-13"),
    (14, "Headphones", 2500.0, "2025-01-14"),
    (15, "Laptop", 60000.0, "2025-01-15"),
    (16, "Keyboard", 1500.0, "2025-01-16"),
    (17, "Mouse", 500.0, "2025-01-17"),
    (18, "Tablet", None, "2025-01-18"),        # Null amount
    (19, "Printer", 8000.0, "2025-01-19"),
    (20, "Monitor", 12000.0, "2025-01-20"),
    
    # Duplicate records
    (2, "Mouse", 500.0, "2025-01-02"),
    (5, "Printer", 8000.0, "2025-01-05"),
    (7, "Laptop", 55000.0, "2025-01-07"),
    
    # Additional records
    (21, "Webcam", 2200.0, "2025-01-21"),
    (22, "Microphone", 3200.0, "2025-01-22"),
    (23, "SSD", 7000.0, "2025-01-23"),
    (24, "RAM", 4500.0, "2025-01-24"),
    (25, None, None, None),                    # All nulls except ID
    (26, "Graphics Card", 45000.0, "2025-01-26"),
    (27, "UPS", 6500.0, "2025-01-27")
]

# Define schema
schema = StructType([
    StructField("sales_id", IntegerType(), True),
    StructField("item", StringType(), True),
    StructField("amount", DoubleType(), True),
    StructField("sale_date", StringType(), True)
])

# Create DataFrame
sales_df = spark.createDataFrame(sales_data, schema=schema)

# Show DataFrame
sales_df.show(truncate=False)

# Print schema
sales_df.printSchema()
```
